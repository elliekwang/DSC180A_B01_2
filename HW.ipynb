```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# Assuming your data is in a DataFrame called `outflows`
# and you have a column 'prism_consumer_id' and 'amount'

results = []

for i in range(10):
    # Step 1: Split at consumer level
    consumers = outflows['prism_consumer_id'].unique()
    train_consumers, test_consumers = train_test_split(consumers, test_size=0.2, random_state=i)
    
    # Step 2: Create train/test DataFrames
    out_train_df = outflows[outflows['prism_consumer_id'].isin(train_consumers)]
    out_test_df = outflows[outflows['prism_consumer_id'].isin(test_consumers)]
    
    # Step 3: Compute mean amount
    train_mean = out_train_df['amount'].mean()
    test_mean = out_test_df['amount'].mean()
    
    # Step 4: Store results
    results.append({
        'run': i+1,
        'train_mean': train_mean,
        'test_mean': test_mean,
        'test_greater_than_train': test_mean > train_mean
    })

# Convert results to DataFrame for easy viewing
results_df = pd.DataFrame(results)

# Print summary
print(results_df)
print("\nTest mean > Train mean count:", results_df['test_greater_than_train'].sum(), "out of 10 runs")
```
